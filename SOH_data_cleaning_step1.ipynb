```python
moving=data.drop(data[(data['t_current']<=0)&(data['speed']==0)].index) #停车or停车充电，要删除
```

```python
moving=moving.drop(moving[moving['t_current']<0].index) #制动充电,删除，最终保留放电数据
```

```python
moving
```

```python
moving['vid'].value_counts()
```

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(n_estimators=100,n_jobs=-1,verbose=2)          #异常值筛选，孤立森林算法（不太明白这个算法，参考论文用的）           
```

```python
moving=moving.fillna(0)
```

```python
X_cols=["speed","mileage","t_volt","t_current","soc","max_temp","min_temp"]
```

```python
model.fit(moving[X_cols])
```

```python
shape=moving.shape[0]
```

```python
batch=10**6
```

```python
all_pred = []
for i in range(int(shape/batch+1)):
    start = i * batch
    end = (i+1) * batch
    test = moving[X_cols][start:end]
    # 预测
    pred = model.predict(test)
    all_pred.extend(pred)
```

```python
moving['pred']=all_pred
```

```python
moving.head()
```

```python
moving['pred'].value_counts()
```

```python
moving_normal=moving.drop(moving[moving['pred']==-1].index)  #删除了异常值，这一步不确定性比较大， 不知道异常值筛选的对不对
```
